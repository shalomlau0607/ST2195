```python
from sklearn.metrics import plot_roc_curve
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier  # added classification model
from sklearn.gaussian_process.kernels import RBF # added classification model
from sklearn.svm import LinearSVC  # added classification model
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer #transform different types
```

```python
##Pre-processing stage
```

```python
###for numerical features
```

```python
numerical_features = ['DepDelay', 'ArrDelay']

#applying SimpleImputer and StandardScaler into a pipeline
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer()),
    ('scaler', StandardScaler())])
numerical_features
numerical_transformer
```

```python
##for categorical features
```

```python
categorical_features = ['FlightNum', 'Month', 'DayOfMonth']

#applying SimpleImputer and then OneHotEncoder into another pipeline
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer()),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])
categorical_features
categorical_transformer
```

```python
data_transformer = ColumnTransformer(
    transformers=[
        ('numerical', numerical_transformer, numerical_features),
        ('categorical', categorical_transformer, categorical_features)])
data_transformer
```

```python
##merged pipeline now contains all the pre-processing steps, and can be used as starting point for different learners
```

```python
#Now ready to complete machine learning pipeline by attaching a learner to the pipeline from above
pipe_lr = Pipeline(steps=[('data_transformer', data_transformer),
                         ('pipe_lr', LogisticRegression(max_iter=1000000))])
pipe_lr
```

```python
def get_delay_counts(df):
    grp = df.groupby(['DepDelay', 'ArrDelay'])['DayofMonth', 'Month'].nunique()
    return {key: grp[key] for key in list(grp.keys())}
def get_delay_proportions(df):
    delay_counts = get_delay_counts(df)
    return {val[0]: round(val[1]/df.shape[0], 4) for val in delay_counts.items()}

print("Dataset delay counts", get_delay_counts(df))
print("Dataset delay proportions", get_delay_proportions(df))
```

```python
##split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df, 
                                                    test_size=0.5, 
                                                    random_state=1)

```

```python
##to tune the pipeline, we can use grid search cross-validation using GridSearchCV()
```

```python
#hyperparameters are set using separated parameter names
param_grid = {
    'data_transformer__numerical__imputer__strategy': ['mean', 'median'],
    'data_transformer__categorical__imputer__syrategy': ['constant', 'most_frequent']
}

grid_lr = GridSearchCV(pipe_lr, param_grid=param_grid)

grid_lr.fit(X_train, y_train);
```
